```
def combine_list(x,y,z):                                        
  for name,age,city in zip(x,y,z):
    print(name, " age ", str(age), ", lives in ", city)             # Zip combines lists into one complete list
                                                                    # [1,2] and [3,4] would be [(1,3) , (2,4)]
name = ['Vinod', 'Harsh', 'Preet', 'David', 'Johan']
age = [17,19,21,19,22]
city = ['Mumbai', 'Delhi', 'Delhi', 'New York', 'Berlin']

combine_list(name,age,city)
```

```
def func(x):                                         # Here we create a function that returns a mathematical equation
  return (x**2 + 5*x)

lst = [1,2,4,7,9]                                    # If we wanted to apply each number in this list to the function above, we would use a FOR loop and a new empty list
lstResult = []
for i in lst:
  lstResult.append(func(i))
print("Using def and for loop:", lstResult)

shorterVersion = lambda x: x**2 + 5*x                # But with lambda, we can create a one line function. you simply define the input, x in this case, and then after the colon
result = list(map(shorterVersion, lst))              # you define the mathmatical equation, much shorter way of defining a function in python
print("Using lambda and map:", result)               # Now we create 'result' that uses map to apply the list of numbers to the function. It is same as using for loop             
                                                     # map(fuction, list) is how it is used. No need to create new list or append.
  
num = [1.32,0.87,1.44,7,0.4,-3,0.6,-7.7,5]
lst2 = []                                            # Instead of using FOR loops to find numbers greater than 1 for example, we can use filter function
for i in num:                                        # It works same way as map but it returns a boolean and creates a new list with items that passed as "True"
  if i > 1:                                          
    lst2.append(i)
print("")
print("Using FOR loop:", lst2)

lst_filter = list(filter(lambda x: x > 1, num)) 
print("Using filter:", lst_filter)
```

```
import re                                            # Import regular expressions module. RE helps us parse strings in many meaningful ways 
                                                    
def find(string):                                
  find_all_names = re.findall(r'[A-Z]{1}[a-z]+', string)         # [A-Z] == any uppercase letter | {1} == amount of letters to look for 
  find_all_ages  = re.findall(r'\d+', string)                    # [a-z] == any lowercase letter |  +  == look for continuous lowercase letters
  print("Names in string:", find_all_names)
  print("Ages in string:", find_all_ages)
  
find("Preet (17 years old) lives in Delhi, India with her dad Yuvraj, a proud Indian army veteran of age 48")
```

```
def find_date(string):
  date = re.findall(r'[A-Z][a-z]+\s+\d+\W{2}\d+', string)         # Date format example: January 1, 2001
  return date

date_string = "On December 31, 1999 the whole world stood still. The year 2000 had come. The next day, the whole world woke up to a new year. On the calendars it said: January 1, 2000."
print("Dates found in string:", find_date(date_string))
```

    [A-Z][a-z]+ == first look for a uppercase letter followed by as many lowercase letters
 
    \s+\d+      == then look for as many whitespace (s) followed by as many digits 
    
    \W{2}\d+    == finally look for two non number/letter (since we need to pass the comma and the whitespace between day and year) then look for as many digits

```
class MyClass:                                   # Capitalize the name of your class
  def __init__(self, string):                    # This is initialization or constructor in Java. Use __init__ to initialize and always add self as parameter and when defining variables.
    self.start = string                          # This is called an attribute (AKA variable) It is useful when adding many values without manually creating a variable for each of them
    
  def prepend(self, string):                     # This is called a method, like in Java. A method is simply a function within a class. Self as parameter is always required
    print(self.start, string)                    
    
instance = MyClass("Prepended string:")          # This is called an instance of MyClass class. It adds new variables to the class which we can use to call many functions of a class.
instance.prepend("Hi there")                     # Prepend is a function/method of MyClass.

instance2 = MyClass("Another prepended string:") # We create another instance using same method. 
instance2.prepend("Hello")
```
